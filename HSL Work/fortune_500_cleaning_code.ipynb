```python
def make_dict(junk):
    diamond = dict()
    dumpster  = junk.split("], [")
    for i,trash in enumerate(dumpster):
        rubbish = trash.split("<', '>")
        if len(rubbish) == 1:
            if i == 0: 
                key = rubbish[0][4:-2]
            elif i == len(dumpster) - 1:
                key = rubbish[0][2:-4]
            else:
                key = rubbish[0][2:-2]
            diamond[key] = None
        else:
            if i == 0:
                key = rubbish[0][4:]
                value = rubbish[-1][:-2]
                diamond[key] = value
            elif i == len(dumpster) - 1:
                key = rubbish[0][2:]
                value = rubbish[-1][:-4]
                diamond[key] = value
            else:
                if len(rubbish) == 3:
                    key = rubbish[0][2:]
                    value = rubbish[1]
                    diamond[key] = value
                    key2 = key + "Growth"
                    value2 = rubbish[-1][:-2]
                    diamond[key2] = value2
                else:
                    key = rubbish[0][2:]
                    value = rubbish[-1][:-2]
                    diamond[key] = value                
            
    return diamond
```

```python
test_df = pd.DataFrame(df["ceos_page_data"].apply(make_dict).values.tolist())
cleaned_df = df.join(test_df).drop(columns = ["ceos_page_data","ceo_name","","Unnamed: 0","index"])
cleaned_df.to_csv("Cleaned_F500data.csv")
```
