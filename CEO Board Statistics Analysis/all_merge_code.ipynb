```python
import pandas as pd
import numpy as np
import scipy as sc
import re
```

```python
df1 = pd.read_csv("Cleaned_F500data.csv")
df2 = pd.read_csv("wsjS&P.csv")
df3 = pd.read_csv('Cleaned_ceo_rank_glassdoor_data.csv')
df4 = pd.read_csv('Cleaned_ceo_rank_hbr.csv')
```

```python
#Renaming to match columns names for the joining company names
df1 = df1.rename(columns = {"companies":"Company"})
df3 = df3.rename(columns = {"GD_company": "Company","CEO Rank": "GD CEO Rank"})
df3 = df3.iloc[:,1:]
df4 = df4.rename(columns = {"HBR_Company": "Company","Rank": "HBR CEO Rank"})
df4 = df4.iloc[:,1:]
test3 = df3.copy()
test4 = df4.copy()
test2 = df2.copy()
test1 = df1.copy()
```

```python
df1.columns
```

```python
clean_amp = lambda text: re.sub(r"&amp;","&",text)
clean_space = lambda text: text[:-1] if text[-1] == " " else text
clean_period = lambda text: re.sub(r"\."," ",text)
clean_white = lambda text: re.sub(r" ","",text)
clean_company = lambda text: re.sub(r"company$","",text)
clean_companies = lambda text: re.sub(r"companies$","",text)
clean_group = lambda text: re.sub(r"group$","",text)
clean_corp = lambda text: re.sub(r"corporation","corp",text)
lowercase = lambda text: str.lower(text)
clean_funcs = [clean_amp,clean_space,clean_period,clean_white,clean_company,clean_companies,clean_group,lowercase]
```

```python
for func in clean_funcs:
    test2["Company"] = test2["Company"].apply(func)
    test1["Company"] = test1["Company"].apply(func)
    test4["Company"] = test4["Company"].apply(func)
    test3["Company"] = test3["Company"].apply(func)
```

```python
#join on company by inner join
wsj_F500_merged = test1.merge(test2,on = "Company")
wsj_F500_merged.to_csv("wsj_F500_merged.csv")
```

```python
wsj_F500_merged.shape
```

```python
#join on company by left join
wsj_F500_merged1 = test1.merge(test2,on = "Company",how='left')
wsj_F500_merged1.to_csv("wsj_F500_merged1.csv")
```

```python
#join on company by outer join
wsj_F500_merged1 = test1.merge(test2,on = "Company",how='outer')
wsj_F500_merged1.to_csv("wsj_F500_merged2.csv")
```

```python
#join ceo ranked companies by outer join
ceo_merged1 = test3.merge(test4,on='Company',how='outer')
ceo_merged1.to_csv('ceo_merged1.csv')
```

```python
## Merge all files together, get rid of unnamed columns
all_merged = wsj_F500_merged1.merge(ceo_merged1,on='Company',how='outer')
all_merged = all_merged.iloc[:,1:]
all_merged.to_csv('merged_dataset.csv')
```
