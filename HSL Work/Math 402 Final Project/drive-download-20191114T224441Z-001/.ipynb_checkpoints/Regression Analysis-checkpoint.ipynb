```python
import numpy as np
from scipy import stats
from scipy import linalg
from matplotlib import pyplot as plt
import pandas as pd
import time
import statsmodels.api as sm
```

```python
data = pd.read_csv('dataproject_interest_full.csv')
```

```python
data
```

```python
sectors = ['Aerospace & Defense', 'Apparel', 'Business Services', 'Chemicals',
       'Energy', 'Engineering & Construction', 'Financials',
       'Food & Drug Stores', 'Food, Beverages & Tobacco', 'Health Care',
       'Hotels, Restaurants & Leisure', 'Household Products', 'Industrials',
       'Materials', 'Media', 'Motor Vehicles & Parts', 'Retailing',
       'Technology', 'Telecommunications', 'Transportation', 'Wholesalers']
divisions = ['East North Central', 'East South Central', 'Middle Atlantic',
       'Mountain', 'New England', 'Pacific', 'South Atlantic',
       'West North Central', 'West South Central']
market_sizes = list(data.filter(regex='^MarketSize',axis=1).columns.unique())
controls = sectors[1:] + divisions[1:] + market_sizes[1:]
```

```python
data[sectors + divisions]
```

```python
data.columns
```

# Profits

```python
interested = ['HBR Ranked','GD Ranked', 'GD & HBR Ranked','Median_pay','Median_age','women_on_board','No_Directors','Board_Independance']
for i in range(3,9):
    print(i)
    regdata = data[['Profits ($M)' ] + interested[:i] + controls].dropna()
    X = regdata[interested[:i] +controls]
    Y = regdata['Profits ($M)']

    results = sm.OLS(Y,sm.add_constant(X)).fit()
    print(results.summary())
```

# Reverse Causality

```python
controls = ['Assets ($M)'] + sectors[1:] + divisions[1:] + market_sizes[1:]
```

```python
interested = ['Profits ($M)','Revenues ($M)','women_on_board','No_Directors','Board_Independance']
for i in range(2,6):
    print(i)
    regdata = data[['HBR Ranked'] + interested[:i] + controls].dropna()
    X = regdata[interested[:i] + controls]
    Y = regdata['HBR Ranked']

    results = sm.Logit(Y,sm.add_constant(X)).fit()
    print(results.summary())
```

```python
interested = ['Profits ($M)','Revenues ($M)','women_on_board','No_Directors','Board_Independance']
for i in range(2,6):
    print(i)
    regdata = data[['GD Ranked'] + interested[:i] + controls].dropna()
    X = regdata[interested[:i] + controls]
    Y = regdata['GD Ranked']

    results = sm.Logit(Y,sm.add_constant(X)).fit()
    print(results.summary())
```
